# DATA DICTIONARY:-

1.__id :-__ The course ID of that particular course. <br>
2.__title:-__ Shows the unique names of the courses available under the development category on ed-tech platform.<br>
3.__url :-__ Gives the URL of the course.<br>
4.__is_paid :-__ Returns a boolean value displaying true if the course is paid and false if otherwise.<br>
5.__num_subscribers :-__ Shows the number of people who have subscribed that course.<br>
6.__avg_rating :-__ Shows the average rating of the course.<br>
7.__avg rating :-__ recent Reflects the recent changes in the average rating.<br>
8.__num_reviews :-__ Gives us an idea related to the number of ratings that a course has received.<br>
9.__num_ published_lectures:-__ Shows the number of lectures the course offers.<br>
10.__num_ published_practice_tests :-__ Gives an idea of the number of practice tests that a course offers.<br>
11.__created :-__ The time of creation of the course.<br>
12.__published_time :-__ Time of publishing the course.<br>
13.__discounted_ price_amount :-__ The discounted price which a certain course is being offered at.<br>
14.__discounted_ price_currency :-__ The currency corresponding to the discounted price which a certain course is being offered at.<br>
15.__price_ detail_amount :-__ The original price of a particular course.<br>
16.__price_ detail_currency :-__ The currency corresponding to the price detail amount for a course.<br>
17.__rating :-__ Shows course rating.<br>
18.__num_reviews :-__ Number of reviews a course received.<br>
19.__is_wishlisted :-__ Shows if course is wishlisted.<br>

```python
#Libaries Required
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.reset_option('max_columns')
import warnings
warnings.filterwarnings("ignore")
```

```python
#loading the dataset
df=pd.read_csv("online_All_Business_p1_p626.csv")
```

```python
df.head(2)
```

```python
#checking for the shape of the data
df.shape
```

```python
df.info()
```

```python
#checking for the column descriptions
df.describe()
```

```python
#picking up the columns that are important for us only
df.drop(columns=['url','avg_rating_recent','published_time','discount_price__currency','discount_price__price_string','price_detail__price_string','price_detail__currency'],inplace=True)
```

```python
df.head(2)
```

```python
#checking for null values
df.isnull().sum()
```

```python
#treating the null-values
```

```python
#discount_price__amount
#over here we agonna create an extra column and fill it with mean and median and then check the distribution with the actual columns
df['discount_price__amount_mean']=df['discount_price__amount'].fillna(df['discount_price__amount'].mean())
df['discount_price__amount_median']=df['discount_price__amount'].fillna(df['discount_price__amount'].median())
```

```python
#ploting the ditribution of the new columns with the actual columns [With mean]
print(df['discount_price__amount'].std())
print(df['discount_price__amount_mean'].std())

fig = plt.figure()
ax = fig.add_subplot(111)
df['discount_price__amount'].plot(kind='kde', ax=ax)
df['discount_price__amount_mean'].plot(kind='kde', ax=ax, color='green')
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
plt.show()
```

```python
#ploting the ditribution of the new columns with the actual columns [With meadian]
print(df['discount_price__amount'].std())
print(df['discount_price__amount_median'].std())

fig = plt.figure()
ax = fig.add_subplot(111)
df['discount_price__amount'].plot(kind='kde', ax=ax)
df['discount_price__amount_median'].plot(kind='kde', ax=ax, color='green')
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
plt.show()
```

As, we can see that both the distribution are same if we fill the null values either with mean and median, so we are gonna choose 
to feel the null values with the meadian over mean as median will take the middle value and we will be dropping those extra columns
created

```python
df['discount_price__amount']=df['discount_price__amount'].fillna(df['discount_price__amount'].median())
```

```python
df.drop(columns=['discount_price__amount_mean','discount_price__amount_median'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(2)
```

```python
df.info()
```

```python
#Replacing the created column values so that we can change it to date-time format
df['created']=df['created'].str.replace('T','')
df['created']=df['created'].str.replace('Z','')
```

```python
df['create_date']=df['created'].str[:10]
```

```python
df['create_date']=pd.to_datetime(df['create_date'])
```

```python
df['create_date_day']=pd.to_datetime(df['create_date']).dt.day
df['create_date_month']=pd.to_datetime(df['create_date']).dt.month
df['create_date_year']=pd.to_datetime(df['create_date']).dt.year
```

```python
df.head(2)
```

```python
df.drop('created',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
#Checki for duplicated records
df[df.duplicated()]
```
